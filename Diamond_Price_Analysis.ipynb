From the above description of the data, the analysis are:


*   Numerical Columns : Carat,clarity, depth, tablem x, y, z, price
*   Dependent variable : Price || Independent Variable : all others except price



  

```python
# here in this data the "Unnamed" column is of no use and relevance. So, we drop it
new_cubic = cubic.drop('Unnamed: 0', axis=1)
```

```python
new_cubic.head()
```

```python
new_cubic.describe(include = ['objec t'])
```

```python
# number of rows that are containing vaues : 0
print("Nu. of rows in carat containing 0: ", format((new_cubic.carat == 0).sum()))
print("No. of rows in depth containing 0's:", format((new_cubic.depth == 0).sum()))
print("No. of rows in x containing 0's: ", format((new_cubic.x == 0).sum()))
print("No. of rows in y containing 0's: ", format((new_cubic.y == 0).sum()))
print("No. of rows in z containing 0's: ", format((new_cubic.z == 0).sum()))
```

```python
# removing the diamonds that have x = y= z = 0, becasue they are dimensionless
new_cubic = new_cubic.drop(new_cubic[new_cubic["x"] == 0].index)
new_cubic = new_cubic.drop(new_cubic[new_cubic["y"] == 0].index)
new_cubic = new_cubic.drop(new_cubic[new_cubic["z"] == 0].index)
new_cubic.shape
```

Earlier the shape of Data was : (26967, 11)


---

After removing the dimensionless diamonds the shape of rhe data is : (26958, 10)

```python
# checkig to duplicates and null values
duplicate_values = new_cubic.duplicated()
print("No. of duplicate values are: ", duplicate_values.sum())
```

```python
new_cubic = new_cubic.drop_duplicates()
```

** multivariate Analysis **

```python
new_cubic.head()
```

```python
# using seaborn library for multivariate analysis
columns = ["carat", "clarity", "depth", "table", "x", "y", "z", "price"]
columns_cubic = (new_cubic[columns])

# usisng pairplot for the representation
sns.pairplot(columns_cubic, diag_kind='kde')  
plt.show()
```

```python
# heat map

plt.figure(figsize=(10,10))
sns.heatmap(new_cubic.corr(), fmt="d")
plt.show()
```

```python
# since the values are duplicating in the heat map, so for better understanding we remove the values above the diaognal

plt.figure(figsize=(10,10))
sns.heatmap(new_cubic.corr(), mask = np.triu(new_cubic.corr(),+1,))
plt.show()
```
